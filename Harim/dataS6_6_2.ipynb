# Pandas basic

# DataFrame

```python
import pandas as pd
```

```python
# 공식문서(도움말) 확인하기
# pd.DataFrame?

df = pd.DataFrame(
{"a" : [4, 5, 6,4],
"b" : [7, 8, 9,9],
"c" : [10, 11, 12,10]},
index = [1, 2, 3,4])
df
```

```python
df["a"]

```

# Subset

```python
df[df["a"] >4]
```

```python
# df["a", "b"]
# 두 개 이상의 값을 불러올 때 Series형태로 불러올 경우 키값 오류가 발생합니다.

df[["a", "b"]]
# DataFrame 형태로 불러와야 합니다.
```

# Reshaping
## sort_values,drop

```python
# a col 기준으로 정렬해줌

df["a"].sort_values()
```

```python
df.sort_values("a",ascending=False)
```

```python
df=df.drop(["c"],axis =1)
# col 기준으로 drop axis=1
df
```

# Group Data

# Groupby, pivot_table

```python
# a col로 grouby 해서 b라는 col 값으로 평균을 구할 수 있다.
df.groupby(["a"]) ["b"].agg(["mean","sum","count"])
```

```python
df.groupby(["a"]) ["b"].describe()
```

```python
# 피보테이블 구하기 

pd.pivot_table(df, index="a",values="b",aggfunc="sum")

# aggfunc = "sum" -> 합 구하기
```
